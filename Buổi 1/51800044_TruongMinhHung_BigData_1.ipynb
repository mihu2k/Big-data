```python
import pyspark
from pyspark import SparkConf, SparkContext
import collections
conf = SparkConf().setMaster("local").setAppName("count")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
text_file = sc.textFile("drive/MyDrive/Big Data/data/small_text.txt")
```

```python
words = []
for line in text_file.collect():
  line = line.lower()
  words.extend(line.split(" "))
rdd = sc.parallelize(words)
# tuple
key = rdd.map(lambda word: (word, 1))
print(key.collect())
```

```python
print(len(words))
```

```python
# count word
counts = key.reduceByKey(lambda x, y: x + y)
print(counts.collect())
```

```python
# tạo ra N số có giá trị từ 1-10
# đếm tần suất xuất hiện của mỗi số
import random
i = random.randint(0, 10)
#print(i)
n = 100
arr = [random.randint(0, 10) for i in range(n)]
#print(arr)

arr = sc.parallelize(arr)
key = arr.map(lambda i:(i, 1))
print(key.collect())

count = key.reduceByKey(lambda x, y: x + y)
print(count.collect())
```
