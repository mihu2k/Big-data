```python
import pyspark
from pyspark import SparkConf, SparkContext
import collections

conf = SparkConf().setMaster("local").setAppName("count")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
# Ví dụ minh họa các method chính
rdd = sc.parallelize([1, 2, 7, 4, 5, 5])
print("Data:", rdd.collect())

# map
map1 = rdd.map(lambda x: (1, x)).collect()

print("Map:", map1)

# filter
filter1 = rdd.filter(lambda item: item > 2).collect()

print("Filter:", filter1)

# distinct
distinct1 = rdd.distinct().collect()

print("Distinct:", distinct1)

# flatMap
flatMap1 = rdd.flatMap(lambda x: (1, x)).collect()

print("Flatmap:", flatMap1)

# sortBy
rddSort = sc.parallelize([('a', 1), ('b', 2), ('1', 3), ('d', 4), ('2', 5), ('2', 5)])
sortBy1 = rddSort.sortBy(lambda x: x[0]).collect() # sort by key
sortBy2 = rddSort.sortBy(lambda x: x[1]).collect() # sort by value

print("Sort by key:", sortBy1)
print("Sort by value:", sortBy2)

# reduceByKey
reduceByKey1 = rddSort.reduceByKey(lambda accum, n: accum + n).collect();
print("Reduce by key:", reduceByKey1)

# collect
print("Collect:", rdd.collect())

# count
print("Count:", rdd.count())

# first
print("First:", rddSort.first())

# reduce
reduce1 = rdd.reduce(lambda accum, item: accum + item)
print("Reduce:", reduce1)

# min, max
min1 = rdd.min()
max1 = rdd.max()

print("Min: {}, Max: {}".format(min1, max1))

# take
print("Take:", rddSort.take(3))
```

```python
# Ứng dụng RDD để thực hiện lại bài toán đếm từ trong một file text 
# lấy ra danh sách 10 từ có tần suất xuất hiện nhiều nhất.

text_file = sc.textFile("drive/MyDrive/Big Data/data/small_text.txt")

print(text_file.collect())
```

```python
words = []

for line in text_file.collect():
  line = line.lower()
  words.extend(line.split(" "))

rdd_bt2 = sc.parallelize(words)

key = rdd_bt2.map(lambda item: (item, 1))

print(key.collect())
```

```python
counts = key.reduceByKey(lambda accum, n: accum + n)
print(counts.collect())
```

```python
result = counts.sortBy(lambda item: item[1], ascending=False).collect()[:10]

print(result)
```
