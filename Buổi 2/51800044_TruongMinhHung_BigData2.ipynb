```python
from pyspark import SparkContext
```

```python
sc = SparkContext(appName = "Test")
```

```python
import random

i = random.randint(0, 10)
print(i)
n =100
arr = [random.randint(0, 10) for i in range(n)]
print(arr)
```

```python
rdd1 = sc.parallelize(arr)
print(rdd1.collect())
```

```python
rdd2 = rdd1.filter(lambda x: x % 2 == 0)
print(rdd2.collect())
```

```python
rdd3 = rdd2.map(lambda x: (x, 1)).reduceByKey(lambda accum, n: accum + n)
print(rdd3.collect())
```

```python
rdd4 = rdd3.sortBy(lambda x: x[1], ascending=False).take(3)
print(rdd4)
```

```python
n1 = 1000
x = [random.randint(0, 100) for i in range(n1)]
y = [random.randint(0, 100) for i in range(n1)]
print(x)
print(y)
point = list(zip(x, y))
print(point)

```

```python
import math
test = (8, 15)


rdd_bt = sc.parallelize(point)
rdd1_bt = rdd_bt.map(lambda item: (item, (math.sqrt((item[0] - test[0])**2 + (item[1] - test[1])**2))))
print(rdd1_bt.collect())

rdd2_bt = rdd1_bt.sortBy(lambda x: x[1], ascending=True)
res = rdd2_bt.map(lambda x: x[0]).take(10)
print(res)
```
